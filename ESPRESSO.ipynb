```python
#Weight Function Calculation

def MIN_WEIGHTS(num):
  list1 = []
  for i in range(len(num)):
    list2 = [] 
    for j in range(no_of_ip):
      list2.append(num[i][j][0])
      list2.append(num[i][j][1])
    list1.append(list2)

  c1 = 1
  c2 = 0
  col_weight = []
  for i in range(len(list1)):
    for j in range(len(list1[0])):
      if (i == 0):
        if(list1[i][j] == 1):
          col_weight.append(c1)
        else:
          col_weight.append(c2)
      else:
        if(list1[i][j] == 1):
          col_weight[j] = col_weight[j] + 1

  k = 0
  impl_weight = []
  for i in range(len(num)):
    wt_sum = 0
    k = 0
    for j in range(no_of_ip):
      wt_sum = wt_sum + (num[i][j][0] * col_weight[k])
      wt_sum = wt_sum + (num[i][j][1] * col_weight[k+1])
      k = k + 2
    impl_weight.append(wt_sum)

  return impl_weight
```

```python
#Soring in Ascending Order
def SORT_IN_ASCEN(impl_weight, pos_list):
  im_wt = impl_weight.copy()
  im_wt.sort()
  im_wt = np.unique(im_wt)
  wtd_ip_list = []
  for i in range(len(im_wt)):
    in1 = [idx for idx, value in enumerate(impl_weight) if value == im_wt[i]]
    for j in range(len(in1)):
      wtd_ip_list.append(in1[j])
  
  new_ip_list = []
  for i in range(len(wtd_ip_list)):
    c1 = wtd_ip_list[i]
    c2 = pos_list[c1]
    #print(c1, " ", c2, '\n')
    new_ip_list.append(c2)
  #print(type(new_ip_list))
  return new_ip_list

#Soring in Ascending Order
def SORT_IN_DESCEN(impl_weight, pos_list):
  im_wt = impl_weight.copy()
  #print(type(im_wt))
  im_wt = np.unique(im_wt)
  im_wt = -np.sort(-im_wt)
  #print(im_wt)
  wtd_ip_list = []
  for i in range(len(im_wt)):
    in1 = [idx for idx, value in enumerate(impl_weight) if value == im_wt[i]]
    for j in range(len(in1)):
      wtd_ip_list.append(in1[j])
  
  new_ip_list = []
  for i in range(len(wtd_ip_list)):
    c1 = wtd_ip_list[i]
    c2 = pos_list[c1]
    #print(c1, " ", c2, '\n')
    new_ip_list.append(c2)
  #print(type(new_ip_list))
  return new_ip_list
```

```python
def valid_or_not (posa, posb):
  #print("func_term: ", posa , '\n', "fbar_term", posb, '\n')
  flag = False
  for i in range(no_of_ip):
    value1 = posa[i][0] and posb[i][0]
    value2 = posa[i][1] and posb[i][1]
    #print("Value 1: ", value1, " Value 2: ", value2)
    if ( value1 == value2 and value1 != True ):
      flag = True
      return flag
      break
  if (flag == False):
    return flag
```

```python
def cons_term_change(posa, posb):
  #print("ENTERED")
  no_chg = 0
  for i in range(no_of_ip):
    value1 = posa[i][0] and posb[i][0]
    value2 = posa[i][1] and posb[i][1]
    #print("Values: ", value1, ' ', value2, '\n')
    if ( value1 == value2 and value1 != True ):
      no_chg += 1
  return no_chg
```

```python
# Expansion done here

def ESPRESSO (np_pos, np_bar_pos, posA, term_count, red_term):
  '''print("Function in ESPRESSO:" , np_pos, '\n')
  print("ESPRESSO PART\n")
  print("Executable Term: ", posA, "\n")'''
  index = []
  for i in range(len(posA)):
    if(i != red_term):
      posA[i][0] , posA[i][1] = posA[i][1] , posA[i][0]
      match = []
      for j in range(len(np_bar_pos)):
        m1 = valid_or_not(posA,np_bar_pos[j])
        match.append(m1)
      #print("Intersection with fbar: ", match, '\n')
      v_flag = False in match
      #print("Intersection with fbar: ", match, " ", v_flag, '\n')
      if(v_flag == False):
        posA[i][0] = True
        posA[i][1] = True
      else:
        posA[i][0] , posA[i][1] = posA[i][1] , posA[i][0]
        index.append(i)
  #print(index)
  #print("Redundant term", posA, '\n')
  new_np_pos, index , term_count_val = IRREDUNDANCY(posA, np_pos, term_count)
  return posA , new_np_pos , index , term_count_val
```

```python
def costFunction(funcA_int):
  costF = 0
  for i in range(len(funcA_int)):
    for j in range(no_of_ip):
      if(funcA_int[i][j] != [1,1]):
        costF += 1
  return costF

def reductionOrder(funcTerm):
  redcF = 0
  input_no = []
  for i in range(len(funcTerm)):
    if ((funcTerm[i][0] == funcTerm[i][1]) and funcTerm[i][0] == True):
      redcF += 1
      input_no.append(i)
  
  ordercF = pow(2, redcF)
  order_list = []
  for i in range(ordercF):
    order_list.append(bin(i)[2:].zfill(redcF))
  return redcF, order_list, input_no
```
